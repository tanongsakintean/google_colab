```python
def text_to_binary(text):
    binary_representation = ''.join(format(ord(char), '08b') for char in text)
    return binary_representation

def binary_to_blocks(binary_str, block_size=32):
    blocks = []
    for i in range(0, len(binary_str), block_size):
        block = binary_str[i:i+block_size]
        # Fill the block to match the block size if it's shorter
        if len(block) < block_size:
            block = block.ljust(block_size, '0')
        blocks.append(int(block, 2))  # Convert binary string to integer
    return blocks

def format_binary_string(binary_str, group_size=8, line_length=8):
    formatted_str = ""
    for i in range(0, len(binary_str), group_size):
        group = binary_str[i:i+group_size]
        formatted_str += group + " "
        # Add a newline after every 'line_length' groups
        if (i // group_size + 1) % line_length == 0:
            formatted_str += "\n"
    return formatted_str.strip()

binary_string = text_to_binary(text)
formatted_binary_string = format_binary_string(binary_string)
message_blocks = binary_to_blocks(binary_string)

# Output
print(f'Binary representation of "{text}":')
print(formatted_binary_string)

print(f'\nMessage blocks (in hexadecimal):')
print([hex(block) for block in message_blocks])
```

```python
def get_k(round):
    if 0 <= round <= 19:
        return 0x5A927999
    elif 20 <= round <= 39:
        return 0x6ed9eba1
    elif 40 <= round <= 59:
        return 0x8f1bbcdc
    else:
        return 0xca62c1d6

def bitwise_operations(a, b, c, d, e, w, round):
    k = get_k(round)
    print(f'====================== t = [{round-1}] ======================')
    print(f'Variables : a={hex(a)}, b={hex(b)}, c={hex(c)}, d={hex(d)}, e={hex(e)}, w[{round-1}]={w[round-1]}, k[{round-1}]={hex(k)}')

    binB = format(b, '032b')
    print(f'B = {binB}')
    shifted_str = binB[30:] + binB[:30]
    binBtoHex = int(shifted_str, 2)
    print(f'= {shifted_str} <-- shift left 30 bits')
    print(f'= {hex(binBtoHex)} = C\n')

    result1 = b & c ^ ~b & d
    print(f'= ({hex(b)} and {hex(c)}) or (not({hex(b)} and {hex(d)}) = {hex(result1)}')

    print(f'               B = {binB}')
    not_binary_str = ''.join('1' if bit == '0' else '0' for bit in binB)
    not_binary_str
    print(f'               NOT B = {not_binary_str} = {hex(int(not_binary_str, 2))}')

    result2 = (result1 + e) % 0x100000000
    print(f'= ({hex(result1)} + {hex(e)}) % {hex(0x100000000)} = ' + hex(result2))

    binA = format(a, '032b')
    print(f'A = {hex(a)} = {binA}')
    shifted_str = binA[5:] + binA[:5]
    binAtoHex = int(shifted_str, 2)
    print(f'               = {shifted_str} = {hex(binAtoHex)}')

    result3 = (result2 + binAtoHex) % 0x100000000
    print(f'= ({hex(result2)} + {hex(binAtoHex)}) % {hex(0x100000000)} = ' + hex(result3))

    result4 = (result3 + int(w[round-1], 16)) % 0x100000000
    print(f'= ({hex(result3)} + {hex(int(w[round-1], 16))}) % {hex(0x100000000)} = ' + hex(result4))

    result5 = (result4 + k) % 0x100000000
    print(f'= ({hex(result4)} + {hex(k)}) % {hex(0x100000000)} = ' + hex(result5))

    a, b, c, d, e = result5, a, binBtoHex, c, d
    print(f't={round}, a={hex(a)}, b={hex(b)}, c={hex(c)}, d={hex(d)}, e={hex(e)}\n')
    return a, b, c, d, e

def process_sha1(rounds, w, a, b, c, d , e):

    for round in range(1, rounds + 1):
        a, b, c, d, e = bitwise_operations(a, b, c, d, e, w, round)

    return a, b, c, d, e

# w = ['0x' + '0' * 8 for _ in range(rounds)]  # Initialize w with placeholder values
final_a, final_b, final_c, final_d, final_e = process_sha1(rounds, w, a, b, c, d , e)

print("Final values:")
print(f"a = {hex(final_a)}")
print(f"b = {hex(final_b)}")
print(f"c = {hex(final_c)}")
print(f"d = {hex(final_d)}")
print(f"e = {hex(final_e)}")
```
