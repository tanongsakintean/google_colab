```python
def text_to_binary(text):
    binary_representation = ''.join(format(ord(char), '08b') for char in text)
    return binary_representation

def binary_to_blocks(binary_str, block_size=32):
    blocks = []
    for i in range(0, len(binary_str), block_size):
        block = binary_str[i:i+block_size]
        # Fill the block to match the block size if it's shorter
        if len(block) < block_size:
            block = block.ljust(block_size, '0')
        blocks.append(int(block, 2))  # Convert binary string to integer
    return blocks

def format_binary_string(binary_str, group_size=8, line_length=8):
    formatted_str = ""
    for i in range(0, len(binary_str), group_size):
        group = binary_str[i:i+group_size]
        formatted_str += group + " "
        # Add a newline after every 'line_length' groups
        if (i // group_size + 1) % line_length == 0:
            formatted_str += "\n"
    return formatted_str.strip()


binary_string = text_to_binary(text)
formatted_binary_string = format_binary_string(binary_string)
message_blocks = binary_to_blocks(binary_string)

# Output
print(f'Binary representation of "{text}":')
print(formatted_binary_string)

print(f'\nMessage blocks (in hexadecimal):')
print([hex(block) for block in message_blocks])
```

```python
def pad_to_sha512(binary_str):
    original_length = len(binary_str)
    # Step 1: Add a single '1' bit
    binary_str += '1'

    # Step 2: Add '0' bits until the length is congruent to 448 modulo 512
    while len(binary_str) % 512 != 448:
        binary_str += '0'

    # Step 3: Append the original length as a 64-bit binary number
    original_length_bits = format(original_length, '064b')
    binary_str += original_length_bits

    return binary_str

def format_binary_string(binary_str, group_size=8, line_length=8):
    formatted_str = ""
    for i in range(0, len(binary_str), group_size):
        group = binary_str[i:i+group_size]
        formatted_str += group + " "
        # Add a newline after every 'line_length' groups
        if (i // group_size + 1) % line_length == 0:
            formatted_str += "\n"
    return formatted_str.strip()

def binary_to_hex_blocks(binary_str):
    hex_blocks = []
    for i in range(0, len(binary_str), 32):
        block = binary_str[i:i+32]
        hex_block = f"0x{int(block, 2):08x}"
        hex_blocks.append(hex_block)
    return hex_blocks

def process_and_display(text):
    # Convert text to binary
    binary_string = ''.join(format(ord(char), '08b') for char in text)

    # Pad the binary string
    padded_binary_string = pad_to_sha512(binary_string)

    # Format the binary string for display
    formatted_binary_string = format_binary_string(padded_binary_string)

    # Convert to hexadecimal blocks
    hex_blocks = binary_to_hex_blocks(padded_binary_string)

    # Output
    print(f'Binary representation of "{text}":')
    print(formatted_binary_string)
    print("\nMessage blocks (in hexadecimal):")
    return hex_blocks

msg_blocks = process_and_display(text)
msg_blocks_int = []
for block in msg_blocks:
  msg_blocks_int.append(int(block, 16))
print(msg_blocks)
```

```python
def process_message_block(a, b, c, d, message, keys, shift_left):
    for i in range(len(message)):
        print(f'=========================== Round {i+1} ===========================')
        print(f'Variables: A={hex(a)}, B={hex(b)}, C={hex(c)}, D={hex(d)}, K[{i+1}]={hex(keys[i])}')
        print(f'Message block [{i}]: {hex(message[i])}')
        print(f'Shift bit left [{i+1}]: {shift_left[i % len(shift_left)]}')

        # Step 1: Calculate the function result (B and C) or (not B and D)
        func_result = (b & c) | (~b & d)
        print(f'({hex(b)} and {hex(c)}) or (not({hex(b)}) and {hex(d)}) = {hex(func_result)}')

        # Step 2: Add the results together with a, message, and key
        temp_result = (func_result + a) % 0x100000000
        print(f'step F = ({hex(func_result)} + {hex(a)}) % {hex(0x100000000)} = {hex(temp_result)}')

        temp_result2 = (temp_result + message[i]) % 0x100000000
        print(f'step M = ({hex(temp_result)} + {hex(message[i])}) % {hex(0x100000000)} = {hex(temp_result2)}')

        temp_result3 = (temp_result2 + keys[i]) % 0x100000000
        print(f'step K = ({hex(temp_result2)} + {hex(keys[i])}) % {hex(0x100000000)} = {hex(temp_result3)}')

        # Step 3: Binary shifting
        binstr = format(temp_result3, '032b')
        shift_amount = shift_left[i % len(shift_left)]
        shifted_str = binstr[shift_amount:] + binstr[:shift_amount]
        shifted_result = int(shifted_str, 2)

        print(f'Binary:  {binstr}')
        print(f'Shifted: {format(shifted_result, "032b")}')

        # Step 4: Add the shifted result to B
        final_result = (shifted_result + b) % 0x100000000
        print(f'new B =  ({hex(shifted_result)} + {hex(b)}) % {hex(0x100000000)} = {hex(final_result)}\n')

        # Update the variables for the next iteration
        a, b, c, d = d, final_result, b, c


    return a, b, c, d

a, b, c, d = process_message_block(a, b, c, d, msg_blocks_int, keys, shift_left)
```
