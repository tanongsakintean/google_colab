```python
def format_binary_string(binary_str, group_size=4, line_length=16):
    formatted_str = ""
    for i in range(0, len(binary_str), group_size):
        group = binary_str[i:i+group_size]
        formatted_str += group + " "
        # Add a newline after every 'line_length' groups
        if (i // group_size + 1) % line_length == 0:
            formatted_str += "\n"
    return formatted_str.strip()

def binary_to_blocks(binary_str, block_size=64):
    blocks = []
    for i in range(0, len(binary_str), block_size):
        block = binary_str[i:i+block_size]
        # if block[i] have length < 64 auto add 0 to cap max 64bit length
        if len(block) < block_size:
            block = block.ljust(block_size, '0')
            blocks.append(block)
        else:
            blocks.append(block)
    return blocks

def hex_to_binary(hex):
    # แปลงเลขฐาน 16 (hexadecimal) เป็น binary 4 บิตต่อกลุ่ม
    binary_representation = ''
    for char in hex:
        binary_representation += binary_table[int(char, 16)]
    return binary_representation

def binary_to_hex(binary_str):
  decimal_representation = int(binary_str, 2)
  hexadecimal_representation = hex(decimal_representation)[2:].upper()
  return hexadecimal_representation

def text_to_hex(text):
  return ''.join(format(ord(char), '02x') for char in text)

def block_to_inital_permutation(blocks):
    blocks_permutation = []
    for block in blocks:
        initial_permutation = [block[i - 1] for i in initial_permutation_table]
        blocks_permutation.append(''.join(initial_permutation))
    return blocks_permutation

def pc_1(binary):
    key_pc_1 = [binary[i - 1] for i in pc_1_table]
    return ''.join(key_pc_1)

def generate_key_round_1(binary,shift_number):
    left_haft = binary[0:28]
    right_haft = binary[28:56]
    c = []
    d = []

    for i in range(16):
        left_haft = left_haft[shift_number[i]:] + left_haft[:shift_number[i]]
        right_haft = right_haft[shift_number[i]:] + right_haft[:shift_number[i]]
        c.append(left_haft)
        d.append(right_haft)
    return c,d

def pc_2(c,d,pc_2_table):
    key = []
    for i in range(16):
        key_pc_2 = [(""+c[i] + d[i])[j - 1] for j in pc_2_table]
        key.append(''.join(key_pc_2))
    return key

def e_bit_selection(block_pumertation_right):
  return "".join(block_pumertation_right[j-1] for j in e_bit_selection_table)


def e_bit_selection_with_key(block_ebit_selection_right,key):
  return "".join(str(int(block_ebit_selection_right[i]) ^ int(key[i])) for i in range(len(block_ebit_selection_right)))

def s_box(block_ebit_selection_right_with_key):
  chunks = [block_ebit_selection_right_with_key[i:i+6] for i in range(0, len(block_ebit_selection_right_with_key), 6)]
  e_bit_table = [e_bit_table_s1,e_bit_table_s2,e_bit_table_s3,e_bit_table_s4,e_bit_table_s5,e_bit_table_s6,e_bit_table_s7,e_bit_table_s8]
  count = 0
  binary_value = ""
  for data in chunks:
    row = data[0] + data[-1]
    col = data[1:-1]
    value = e_bit_table[count][int(row,2)][int(col,2)]
    binary_value += format(value, '04b')
    count += 1
  return binary_value

def p_box(binary_value):
  p_value = [binary_value[i - 1] for i in p_table]
  return ''.join(p_value)

def create_new_block_r(block_p,blocks_pumertation_left):
  return format(int(block_p,2) ^ int(blocks_pumertation_left,2),'032b')

def inverse_permutation(block_pumertation_left,block_pumertation_right):
  # create a single 64-bit block from the two 32-bit halves
  block = block_pumertation_right + block_pumertation_left
  # apply the inverse permutation to the 64-bit block
  inverse_permutation = [block[i - 1] for i in inverse_permutation_table]
  return ''.join(inverse_permutation)
#hex text
hex_text = text_to_hex(text)

#binary text
# binary_text = hex_to_binary(hex_text)
binary_text = '0000000100100011010001010110011110001001101010111100110111101111'

#blocks binary
blocks = binary_to_blocks(binary_text)


#blocks permutation
blocks_permutation = block_to_inital_permutation(blocks)

#key in hex
key_hex = text_to_hex(key)

#key in pc1
key_pc_1 = pc_1(hex_to_binary(key_hex))
# key_pc_1 = pc_1('0001001100110100010101110111100110011011101111001101111111110001')

#key in c,d
c,d = generate_key_round_1(key_pc_1,number_of_left_shift)

#key in pc_2
k = pc_2(c,d,pc_2_table)

#blocks left
blocks_pumertation_left = [block_permutation[0:32] for block_permutation in blocks_permutation]

#blocks right
blocks_pumertation_right = [block_permutation[32:64] for block_permutation in blocks_permutation]

#step flow DES
for i in range(16):
  blocks_ebit_selection_right.append(e_bit_selection(blocks_pumertation_right[i]))
  blocks_ebit_selection_right_with_key.append(e_bit_selection_with_key(blocks_ebit_selection_right[i],k[i]))
  blocks_sbox.append(s_box(blocks_ebit_selection_right_with_key[i]))
  blocks_p.append(p_box(blocks_sbox[i]))
  blocks_pumertation_right.append(create_new_block_r(blocks_p[i],blocks_pumertation_left[i]))
  blocks_pumertation_left.append(blocks_pumertation_right[i])

#step inverse permutation
#blocks_final.append(inverse_permutation(blocks_pumertation_left[16],blocks_pumertation_right[16]))
blocks_final.append(inverse_permutation('01000011010000100011001000110100','00001010010011001101100110010101'))
#step hex
hex_final = binary_to_hex(blocks_final[0])

```
