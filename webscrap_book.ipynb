```python
from urllib.request import urlopen as req
from bs4 import BeautifulSoup as Soup
```

2. Request URL by get data from url https://www.tmd.go.th/province.php?id=37

```python
#url = 'https://www.tmd.go.th/province.php?id=37'
url = 'https://www.naiin.com/category?type_book=best_seller&product_type_id=1'
webopen = req(url)
page_html = webopen.read()
print(page_html) #report in term of bytes
webopen.close()
```

3 Convert page_html to Soup Object

```python
#you must convert bytes to web_page using beutifulsoup
data = Soup(page_html,'html.parser')
print(data)
#หา tag ที่แสดงสินค้าและราคา
#หรือหาดูได้ที่ browser โดยคลิกขวาที่อุณภูมิแล้ว inspect ดู code
```

4. Find Element tag p ('class: 'txt-price')

```python
temp = data.findAll('p',{'class':'txt-price'})
print(temp)
```

```python
#ตัดคำให้เหลือแค่ราคา โดยใช้ .text เพื่อเอาแค่ข้อความอื่น ๆ ออก
result = temp[1].text
print(result)
```

```python
 #ดึงค่าชื่อหนังสือมาด้วย
 itemname = data.findAll('a',{'class':'itemname'})
#  print(itemname[0].text.strip()) ได้เหมือนกัน

 itn = itemname[0].text.replace(' ','') #ลบช่องว่างด้านหน้า
 print(itn)
 print('ชื่อหนังสือ : {} ราคา : {}'.format(itn,result)) #FORMAT
```

```python
#create function
def bestseller(itnid):
  url = 'https://www.priceza.com/b/computers-and-offices/graphic-cards'
  webopen = req(url)
  page_html = webopen.read()
  webopen.close()
  data = Soup(page_html,'html.parser')

  itemprice = data.findAll('div',{'class':'pz-pdb-price'})
  result = itemprice[itnid].text

  itemname = data.findAll('h3',{'class':'pz-pdb_name'})
  itn = itemname[itnid].text.replace(' ','')

  # print('ชื่อหนังสือ : {}   ราคา : {}'.format(itn,result))
  text = 'ชื่อหนังสือ : {}    ราคา : {}'.format(itn,result)
  return  text

# for i in range(1,10):
#   bestseller(i)
```

```python
#ดักจับ error กรณีที่ไม่มี field ที่ระบุไว้ หรือจังหวัดไม่มีข้อมูล
for i in range(25,35):
  try:
    print(bestseller(i))
  except Exception as e:
    print('no result because',e)
```
